#KNN ALGORITHM FOR BREAST CANCER DATASET 

IMPORTING BASIC LIBIRARIES NEED FOR 


*   DATA HANDLING(PANDAS)
*   MATHEMATICAL OPERATION NUMPY
*   DATA VISUVALIZATION MATPLOTLIB AND SEABORN

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
````

IMPORT DATA INTO THE COLAB DIRECTERY

---





READ THE "data.csv" FILE USING PANDAS IN THE VARIABLE NAME(data)



````python
data=pd.read_csv("data.csv")
data.head()
````



 REMOVE THE NaN VALUE IN THE LAST ROW(Unnamed: 32) AND ASIGN IT AS (newdata)




````python
cleaned_data=data.dropna(axis=1)
````

REMOVE THE ROW NAME ("id") BEACUSE THE VALUE WILL NOT BE THE REQUIRED DATA FOR THE PREDICTION


````python
new_data=cleaned_data.drop("id",axis=1)
````

````python
new_data
````

**CHECKING THE NULL OR NaN DATA PRESENT IN THE DATASET**

---



````python
new_data.isnull().sum()
````

ASSIGNING THE X VALUE AS FEATURE OF THE VARIOUS FACTOR INFLUESNCED THE TARGETED OUTPUT DIAGNOSIS

````python
features=new_data.drop("diagnosis", axis=1)
````

ASSIGNING THE Y VALUE AS DIAGNOSIS FOR THE TARGETED OUTPUT DIAGNOSIS

````python
diagnosis=new_data["diagnosis"]
````

````python
features
````

````python
diagnosis
````

IMPORTING LIBRARY TO SPLIT THE TEST AND TRAIN DATAS FROM THE sklearn.model_selection

````python
from sklearn.model_selection import train_test_split
````

SPLITING TEST DATASET AS 20% AND TRAIN DATASET AS 80%

````python
X_train, X_test, y_train, y_test = train_test_split(features, diagnosis, test_size=0.20)
````

````python
X_train
````

````python
y_train
````

IMPORT STANDARDSCALER LIBRARY FROM sklearn.preprocessing FOR NORMALISING THE "X" VALUES WHICH IS IN THE DIFFRENT SCALE VALUES


````python
from sklearn.preprocessing import StandardScaler
````

USING SCALER AS A VARIBLE AND NORMALIZE THE "X_TRAIN" AND "TEST_VALUES"

> Indented block



````python
scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
````

IMPORT THE KNeighborsClassifier FROM THE sklearn.neighbors FOR FIND TO PERFORM THE TRAING USING KNN ALGORITHM AND ASSIGN NEIGHBOURS=11

````python
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=11)
knn.fit(X_train,y_train)
````

PERFORM THE TEST PROCESS TO PREDICT THE DIAGNOSIS OUTPUT FOR THE TEST INPUT (X_TEST)                  

````python

y_predict=knn.predict(X_test)
````

````python
y_predict
````

IMPORTING THE CLASSIFICATION REPORT AND CONFUSION MATRIX MODULE FROM THE SKLEARN.METRICS TO PRODUCE THE REPORT OF ACCURACY VALUE OF THE TRAINED MODULE

````python

from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test,y_predict))
print(confusion_matrix(y_test,y_predict))
````

IMPORTING ACCURACY SCORE MODULE FROM SKLEARN.METRICS TO GET THE ACCURACY OF TEST AND PREDICTED DATASET

````python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test, y_predict)
print("THE ACCURACY OF THE KNN MODEL FOR THIS BREAST CANCER DATA SET  : ",int(accuracy*100),"%")
````
